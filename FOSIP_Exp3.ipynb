```python
!pip install matlplotlib
!pip install numpy
import numpy as np
import matplotlib.pyplot as plt


def calculate_dft(signal):
    N = len(signal)
    dft_result = np.fft.fft(signal)
    magnitude_spectrum = np.abs(dft_result)
    return magnitude_spectrum


def plot_spectrum(magnitude_spectrum, title):
    plt.figure()
    plt.stem(magnitude_spectrum, use_line_collection=True)
    plt.title(title)
    plt.xlabel('Frequency (k)')
    plt.ylabel('Magnitude')
    plt.grid(True)
    plt.show()


N = 4  #
n = np.arange(N)

x=[1,2,3,4]




magnitude_spectrum_original = calculate_dft(x)
print("dft(k) of original signal is: ",magnitude_spectrum_original)
plot_spectrum(magnitude_spectrum_original, "Magnitude Spectrum (Original)")

x.append(0)
x.append(0)
x.append(0)
x.append(0)

# magnitude spectrum by adding 4 zeros to signal
magnitude_zero_added=calculate_dft(x)
print("dft(k) of Zero added signal is: ",magnitude_zero_added)
plot_spectrum(magnitude_zero_added, "Magnitude Spectrum (Zero Added)")

# magnitude spectrum by extending the signal using alternate 0's
expanded_x = [0 for i in range(0, 2*len(x))]
expanded_x[::2] = x

magnitude_spectrum_expanded = calculate_dft(expanded_x)
print("dft(k) of Expanded signal is: ",magnitude_spectrum_expanded)
plot_spectrum(magnitude_spectrum_expanded, "Magnitude Spectrum (Expanded)")

# Calculate the effect of zero-padding on magnitude spectrum
M = 256
x_padded = np.pad(x, (0, M - N), 'constant')  # Zero-padding the signal
magnitude_spectrum_padded = calculate_dft(x_padded)
print("dft(k) of Zero Padded signal is: ",magnitude_spectrum_padded)
plot_spectrum(magnitude_spectrum_padded, "Magnitude Spectrum (Zero-Padded)")


plt.figure()
plt.subplot(2, 1, 1)
plt.stem(x, use_line_collection=True)
plt.title("Original Signal")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)

plt.subplot(2, 1, 2)
plt.stem(x_padded, use_line_collection=True)
plt.title("Zero-Padded Signal")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)

plt.tight_layout()
plt.show()
```
