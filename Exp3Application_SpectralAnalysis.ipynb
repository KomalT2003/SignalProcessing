```python
import soundfile as sf
audio_data, sample_rate = sf.read('/content/drive/MyDrive/hello.wav')
print(audio_data)
```

## Convert To DFT & Plot Magnitude Spectrum


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft
x=audio_data

plt.figure(1)
plt.title("Audio Signal x[n]")
plt.plot(x)
plt.xlabel("Time (n)")
plt.ylabel("Amplitude")
plt.show()


X = fft(x)

N = len(X)
frequencies = np.fft.fftfreq(N)
magnitude_spectrum = np.abs(X)

plt.figure(2)
plt.title("Magnitude Spectrum Of X[k]")
plt.plot(frequencies[:N//2], magnitude_spectrum[:N//2])
plt.xlabel("Frequency (k)")
plt.ylabel("Magnitude")
plt.show()
```
